# Craw

```python
# Truy cập trang web

browser = webdriver.Chrome()

browser.get('https://tuoitre.vn/tin-moi-nhat.htm')

# Lây url tất cả bài viết
# Báo tuổi trẻ
tuoiTrePostList = []
while True:
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    sleep(5)
    # ghi đè nhiều lần
    postList = browser.find_elements(By.XPATH, "//*[@class='box-category-link-title']")
    for post in postList:
        if post not in tuoiTrePostList:
            tuoiTrePostList = tuoiTrePostList + [post.get_attribute("href")]
    print(len(tuoiTrePostList))
    if (len(tuoiTrePostList) >= 500):
        break
     # Tìm và nhấp vào nút "Xem thêm" nếu tồn tại
    try:
        load_more_button = browser.find_element(By.XPATH, "//*[@class='box-viewmore']/a")
        load_more_button.click()
        print('Đã nhấn xem thêm')
        sleep(2)
    except:
        print("Tự load không cần nhấn xem thêm")
```

# Tiền xử lý

```python
def XuLy(bao):
    # with open(r"D:\GPA40\Nam3\KyI\ChinhThucTrenTruong\KhaiPhaDuLieuWeb\stopwords-project\vietnamese-stopwords.txt", "r", encoding="utf-8") as fd:
    #     stop_vietnamese = fd.read().splitlines()
    
    directory_path = f'raw-data\\{bao}'
    i = 1
    for filename in os.listdir(directory_path):
        file_path = file_path = os.path.join(directory_path, filename)  
        with open(file_path, "r", encoding="utf-8") as f:
            text = f.read()
            textFind = text.split()
            # xóa stop words
            # for text in textFind:
            #     if text in stop_vietnamese:
            #         text = ''
            text = ' '.join(textFind)
            # chuyển văn bản thành chữ thường
            text = text.lower()
            # loại bỏ thẻ HTML và url
            textFind = text.split()
            for word in text:
                if word.startswith('http://') or word.startswith('https://') or word.startswith("www."):
                    word = ''
            text = ' '.join(textFind)
            soup = BeautifulSoup(text, 'html.parser')
            for data in soup(['style', 'script']):
                data.decompose()
            text = soup.get_text()
            # xóa dấu câu, chữ số và ký tự đặc biệt
            text = re.sub(r'[^\w\s]','',text)
            # xóa số
            text = re.sub('\d+',' ', text)
            # tách từ
            tokenized_text = ViTokenizer.tokenize(text)
            # ghi file
            path = f'preprocessing-data\\{bao}\\28_Pre{i}.txt'
            with open(path, "w", encoding="utf-8") as file:
                file.write(tokenized_text)
            i = i + 1

XuLy('dantri')

XuLy('tuoitre')
```
